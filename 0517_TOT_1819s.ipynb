```python
player_list = wd.find_elements_by_class_name('playerName')

player_name_0 = []

for i in range(0, len(player_list)):
    name = player_list[i].text
    player_name_0.append(name)

player_name_0
```

### **선수 이름에 있는 라틴어 디코딩**

```python
from unicodedata import normalize, category

player_name = []

for i in player_name_0:
    a = ''.join([c for c in normalize('NFD', i) if category(c) != 'Mn'])
    player_name.append(a)
    
player_name
```

## **선수 링크 긁어오기**

```python
players_link = wd.find_elements_by_tag_name('a')
players_link
```

```python
from collections import OrderedDict
import re

link = []
new_link = []

for i in range(0, len(players_link)):
    a = players_link[i].get_attribute('href')

    link.append(a)
    
for i in range(0, len(link)):
    a = re.findall("h.*players*.*\d*.*overview$", link[i])
    if len(a) != 0:
        new_link.append(a)
        
new_link = sum(new_link, [])

# 중복제거
new_link = list(OrderedDict.fromkeys(new_link))

new_link
```

```python
players_id = []

for i in range(0, len(new_link)):
               a = int(new_link[i].split('/')[4])
               players_id.append(a)

            
# 선수 테이블과 합치기
team_sheet['id'] = players_id

team_sheet.head()
```

### **공격수 스탯 컬럼명 만들기**

```python
# https://www.premierleague.com/match/22720
from bs4 import BeautifulSoup
import urllib.request
import urllib.parse

# web_url에 원하는 웹의 URL을 넣어주시면 됩니다.

fw_players = []

for i in forward_df['link']:
    with urllib.request.urlopen(i) as response:
        html = response.read()
        soup = BeautifulSoup(html, 'html.parser')    
        
        a = soup.find_all('div', 'statsListBlock')
        
        fw_players.append(a)

fw_players
```

```python
fw_stats_01 = []
fw_stats_02 = []

for i in range(0, len(fw_players[0])):
    for j in range(1, len(fw_players[0][i].find_all('span', 'stat'))):
        a = fw_players[0][i].find_all('span', 'stat')[j].text
        b = a.split(' ')
        
        fw_stats_01.append(b)
        
for k in range(0, len(fw_stats_01)):
    c = fw_stats_01[k][:-3]
    fw_stats_02.append(" ".join(c))

fw_stats_02
```

```python
for i in range(0, len(fw_stats_02)):
    forward_df[fw_stats_02[i]] = np.nan

forward_df
```

### **전체 공격수 스탯 추가하기**

#### **아래 xpath_02 바꿔야함**

```python
# test by kwon
import time
# xpath_01 = """//*[@id="mainContent"]/div[3]/div/div/div/div/div/section/div[2]/div[2]"""
# xpath_02 = """//*[@id="mainContent"]/div[3]/div/div/div/div/div/section/div[2]/ul/li[4]"""

match_wd = webdriver.Chrome('c:/Python/API-master/Soccer_Team/files/chromedriver.exe')

stat_list = []
forward_stat = []

for i in forward_df['link']:
    match_wd.get(i)

    time.sleep(3)

#     ### 선수별 2017/18시즌 페이지 들어가기
#     sel_season_01 = match_wd.find_element_by_xpath(xpath_01).click()

#     time.sleep(5)

#     sel_season_02 = match_wd.find_element_by_xpath(xpath_02).click()

    # 스탯 선택
    tag_class = match_wd.find_elements_by_class_name('statsListBlock')

    for i in range(0, len(tag_class)):
        class_text = tag_class[i].text

    # re.findall("\d+", b)

        stat_list.append(re.findall("\d*\.?\d+", class_text))

# 공격수 개인 능력치
for i in range(0, len(forward_df)):
    a = sum(stat_list[4 * i:(4*i)+4], [])
    forward_stat.append(a)

# 공격수 개인 능력치 종합
for i in range(0, len(forward_df)):
    forward_df.iloc[i, 4:] = forward_stat[i]
    
forward_df
```
